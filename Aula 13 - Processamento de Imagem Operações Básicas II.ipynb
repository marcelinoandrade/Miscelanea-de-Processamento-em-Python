```python
#Accessing Image Properties
%matplotlib notebook 
import cv2
import numpy as np
import matplotlib.pyplot as plt

def meanFilter(im,d):
    for i in range(d,im.shape[0]-d):
        for j in range(d,im.shape[1]-d):
            block = im[i-d:i+d+1, j-d:j+d+1]
            m = np.mean(block,dtype=np.float32)
         #   print(i+j)
            im[i][j] = int(m)
    return im
```

```python

img = cv2.imread('Dantas.jpeg')
img = cv2.resize(img,(512,512))
img=img[:,:,2]

fig = plt.subplots(figsize=(10,10))

ax1=plt.subplot(2, 2, 1)  
ax1.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
ax1.set_title('Meu Bisavo - Antonio Pinheiro Dantas')


ax2=plt.subplot(2, 2, 2)  
N=40
imIn=img.copy()
imgF=meanFilter(imIn,N)
ax2.imshow(cv2.cvtColor(imgF, cv2.COLOR_BGR2RGB))


p=100
ax3=plt.subplot(2, 2, 3)  
ax3.imshow(cv2.cvtColor(img[p:p+5,p:p+5], cv2.COLOR_BGR2RGB))

ax4=plt.subplot(2, 2, 4)  
ax4.imshow(cv2.cvtColor(imgF[p:p+5,p:p+5], cv2.COLOR_BGR2RGB))

```

Expansão ou contração uniforme

```python
def resize(im,d):
    imr=im
    for i in range(im.shape[0]):
        for j in range(im.shape[1]):
            imr[int(i/d)][int(j/d)] = im[i][j]
    return imr
```

Reflexão em torno de um eixo

```python
def refleteY(im):
    imr=im
    for i in range(0,im.shape[0]):
        for j in range(0,im.shape[1]):
            imr[-i+im.shape[0]-1][j] = im[i][j]
    return imr

def refleteX(im):
    imr=im
    for i in range(0,im.shape[0]):
        for j in range(0,im.shape[1]):
            imr[i][-j+im.shape[0]-1] = im[i][j]
    return imr


def refleteXY(im):
    imr=im
    for i in range(0,im.shape[0]):
        for j in range(0,im.shape[1]):
            imr[-i+im.shape[0]-1][-j+im.shape[0]-1] = im[i][j]
    return imr


img = cv2.imread('Eu.JPG')
img = cv2.resize(img,(512,512))
img=img[:,:,2]



s=(2*img.shape[0],2*img.shape[1])

imgm=np.zeros(s,dtype=np.uint8)

imgm[:img.shape[0]-1,:img.shape[1]-1]=img[:img.shape[0]-1,:img.shape[1]-1]

img1=imgm.copy()
img2=imgm.copy()
img3=imgm.copy()


fig = plt.subplots(figsize=(10,10))

ax1=plt.subplot(2, 2, 1)  
ax1.imshow(cv2.cvtColor(imgm, cv2.COLOR_BGR2RGB))
ax1.set_title('Eu')

ax2=plt.subplot(2, 2, 2)  
ax2.imshow(cv2.cvtColor(refleteY(img1), cv2.COLOR_BGR2RGB))

ax3=plt.subplot(2, 2, 3)  
ax3.imshow(cv2.cvtColor(refleteX(img2), cv2.COLOR_BGR2RGB))

ax4=plt.subplot(2, 2, 4)  
ax4.imshow(cv2.cvtColor(refleteXY(img3), cv2.COLOR_BGR2RGB))
```
