**Enquadramento (por que é C1):**
- **H3 – Operações em diferentes conjuntos:** contagem por combinações escolhendo de dois grupos distintos.
- **H2 – Padrões numéricos:** decomposição em casos e uso do Princípio Multiplicativo.
- **Conteúdos:** análise combinatória, combinações.

**Modelagem e solução resumida:**
1. Composições possíveis (mínimo 3 cardiologistas): $3C+2N$, $4C+1N$, $5C$.
2. Total de equipes:
$$\Large
\binom{7}{3}\binom{6}{2}+\binom{7}{4}\binom{6}{1}+\binom{7}{5}.
$$

**Em fatoriais:**

$$\Large
\frac{7!}{3!\,4!}\cdot\frac{6!}{2!\,4!}
+\frac{7!}{4!\,3!}\cdot\frac{6!}{1!\,5!}
+\frac{7!}{5!\,2!}.
$$


**Resposta:** **Alternativa E**.


```python
# ENEM Q139 — equipes com no mínimo 3 cardiologistas
import math, itertools
import matplotlib.pyplot as plt

# tamanhos dos grupos
C, N = 7, 6

# contagens por caso
casos = {
    "3C + 2N": math.comb(C, 3) * math.comb(N, 2),
    "4C + 1N": math.comb(C, 4) * math.comb(N, 1),
    "5C + 0N": math.comb(C, 5)
}
total = sum(casos.values())

print("Contagens por caso:", casos)
print("Total de equipes:", total)

# exemplos concretos (primeiros 5 de cada caso) para entendimento
cards = [f"C{i}" for i in range(1, C+1)]
neurs = [f"N{i}" for i in range(1, N+1)]

def exemplos(kc, kn, limite=5):
    it = itertools.islice(
        (tuple(sorted(c))+tuple(sorted(n))
         for c in itertools.combinations(cards, kc)
         for n in itertools.combinations(neurs, kn)),
        limite
    )
    return list(it)

print("\nExemplos 3C+2N:", exemplos(3,2))
print("Exemplos 4C+1N:", exemplos(4,1))
print("Exemplos 5C+0N:", exemplos(5,0))

# gráfico: quantas equipes em cada caso
fig = plt.figure(figsize=(6,4), dpi=120)
x = list(casos.keys())
y = list(casos.values())
plt.bar(x, y)
for xi, yi in zip(x, y):
    plt.text(xi, yi, f"{yi}", ha="center", va="bottom")
plt.title("Formações possíveis por caso (mín. 3 cardiologistas)")
plt.ylabel("Número de equipes")
plt.grid(True, linestyle=":")
plt.tight_layout()
plt.show()
```

Enquadramento (por que é C1):

$\Large \frac{15}{60} $ e $\Large \frac{4}{100} $



H3 – Operações e proporções: ajuste de tempo e resolução algébrica.

Modelagem e solução resumida:

  $$ \Large P(A\cap B)=P(A)\cdot P(B) $$

 $$\Large p=\frac{15-r}{60},\qquad p^2=\frac{4}{100}=\frac{1}{25} $$

 $$\Large \frac{15-r}{60}=\frac{1}{5}\;\Longrightarrow\;15-r=12\;\Longrightarrow\;r=3\ \text{s} $$

Resposta: 3 s.

```python
# ENEM Q140 — modelo simples (Ensino Médio)

import matplotlib.pyplot as plt

T0 = 15   # tempo vermelho atual (s)
C  = 60   # ciclo (s)

# redução alvo (p^2 = 4/100)
r_star = 3
p_star = (T0 - r_star)/C

# curva discreta: r = 0..15
r = list(range(0, 16))
p2 = [((T0 - x)/C)**2 for x in r]

print(f"Redução ótima: r* = {r_star} s")
print(f"Prob. por semáforo: p* = {p_star:.2f}")
print(f"Prob. dos dois vermelhos: (p*)^2 = {p_star**2:.2f}")

plt.figure(figsize=(6,4))
plt.plot(r, p2)
plt.axhline(0.04, linestyle="--")       # alvo 4/100
plt.axvline(r_star, linestyle=":")      # r* = 3 s
plt.scatter([r_star], [0.04], s=60)     # ponto de solução
plt.title("Probabilidade de dois sinais vermelhos vs redução r")
plt.xlabel("Redução r no tempo vermelho (s)")
plt.ylabel("p(r)^2")
plt.grid(True, linestyle=":")
plt.tight_layout()
plt.show()
```

**Enquadramento (por que é C1):**
- **H3 – Operações e proporções:** razão densidade = habitantes/área, comparação entre regiões.
- **H1 – Representações racionais:** uso de frações $ \tfrac{1}{2} $ e $ \tfrac{3}{4} $.

**Modelagem e solução resumida:**
1. Definição: $ d(X)=\dfrac{H(X)}{A(X)} $.
2. Dados: $ A(Q)=\tfrac{3}{4}A(R) $ e $ H(Q)=\tfrac{1}{2}H(R) $.
3. Cálculo:
   $
   d(Q)=\frac{H(Q)}{A(Q)}
        =\frac{\tfrac{1}{2}H(R)}{\tfrac{3}{4}A(R)}
        =\frac{1/2}{3/4}\cdot\frac{H(R)}{A(R)}
        =\frac{2}{3}\,d(R).
   $

**Resposta:** $\boxed{\,d(Q)=\tfrac{2}{3}\,d(R)\,}$.


```python
# ENEM Q141 — densidade demográfica: d(Q) = (2/3)·d(R)
import matplotlib.pyplot as plt

# Escolha didática: mantém as razões do enunciado
A_R = 1000.0        # área de R (km²)
H_R = 100000.0    # habitantes de R

A_Q = (3/4)*A_R    # área de Q = 3/4 da área de R
H_Q = (1/2)*H_R    # habitantes de Q = 1/2 dos habitantes de R

d_R = H_R / A_R
d_Q = H_Q / A_Q    # deve dar (2/3)*d_R

def anota_barras(ax):
    for b in ax.patches:
        y = b.get_height()
        ax.annotate(f"{y:,.0f}".replace(",", "."),
                    (b.get_x()+b.get_width()/2, y), ha="center", va="bottom", fontsize=9)

# Gráfico 1 — Áreas
plt.figure(figsize=(6,4))
plt.bar(["R","Q"], [A_R, A_Q])
plt.title("Áreas (km²) — Q = 3/4 de R")
plt.ylabel("km²")
plt.grid(True, linestyle=":")
anota_barras(plt.gca())
plt.tight_layout()
plt.show()

# Gráfico 2 — Habitantes
plt.figure(figsize=(6,4))
plt.bar(["R","Q"], [H_R, H_Q])
plt.title("Habitantes — Q = 1/2 de R")
plt.ylabel("habitantes")
plt.grid(True, linestyle=":")
anota_barras(plt.gca())
plt.tight_layout()
plt.show()

# Gráfico 3 — Densidade
plt.figure(figsize=(6,4))
plt.bar(["d(R)","d(Q)"], [d_R, d_Q])
plt.title("Densidade demográfica (hab/km²)")
plt.ylabel("hab/km²")
plt.grid(True, linestyle=":")
anota_barras(plt.gca())
plt.text(0.5, (d_R+d_Q)/2, "d(Q) = 2/3 · d(R)", ha="center")
plt.tight_layout()
plt.show()

print(f"d(R) = {d_R:,.0f} hab/km²".replace(",", "."))
print(f"d(Q) = {d_Q:,.0f} hab/km²".replace(",", "."))
print(f"Razão d(Q)/d(R) = {d_Q/d_R:.2f}")  # ~ 0,67 = 2/3
```

```python
# ENEM Q142 — Diferença entre 3 1/2" e 4 5/6" (visual e numérico)
from fractions import Fraction
import matplotlib.pyplot as plt

# valores exatos como frações
cel1 = Fraction(7, 2)   # 3 1/2
cel2 = Fraction(29, 6)  # 4 5/6
dif  = abs(cel2 - cel1)

# para o gráfico (decimais)
v1 = float(cel1)
v2 = float(cel2)
vd = float(dif)
vd_round = round(vd, 1)

print(f"Celular 1: {cel1} = {v1:.3f} pol")
print(f"Celular 2: {cel2} = {v2:.3f} pol")
print(f"Diferença exata: {dif} = {vd:.3f} pol  →  com 1 casa: {vd_round:.1f} pol")

# gráfico de barras com anotação da diferença
fig, ax = plt.subplots(figsize=(6,4), dpi=120)
ax.bar(["Celular 1", "Celular 2"], [v1, v2])
ax.set_ylabel("polegadas")
ax.set_title("Tamanhos de tela e diferença (em polegadas)")
ax.grid(True, linestyle=":")

# valores numéricos no topo
for x, y in zip([0, 1], [v1, v2]):
    ax.annotate(f"{y:.3f}",
                xy=(x, y), xytext=(0, 3),
                textcoords="offset points", ha="center")

# seta dupla mostrando a diferença no segundo aparelho
x0 = 1.05
ax.annotate("diferença",
            xy=(x0, v1), xytext=(x0, v2),
            arrowprops=dict(arrowstyle="<->"),
            va="center")
ax.text(x0+0.02, (v1+v2)/2, f"{vd:.3f} ≈ {vd_round:.1f}",
        va="center")

plt.tight_layout()
plt.show()
```

**Enquadramento (por que é C1):**
- **H3 – Operações e porcentagens:** parcelas crescem por acréscimo percentual fixo (PG).
- **H2 – Padrões/modularidade:** escolha do mês vira congruência módulo \(12\).

**Modelagem e solução resumida:**
1. Parcela $j$: $P_j=P_1\,q^{\,j-1}$ com $q>1$.
2. Compra no mês $m$. Parcelas pagas pela empresa (maio $=5$) satisfazem  
   $m+j-1\equiv 5 \pmod{12}\ \Rightarrow\ j\equiv 6-m \pmod{12}$.
3. Teste das opções:
   - $m=2$ (fev.): $j\equiv 4$ → $4,16,\dots,100$.
   - $m=4$ (abr.): $j\equiv 2$ → $2,14,\dots,98$.
   - $m=5$ (mai.): $j\equiv 1$ → $1,13,\dots,97$.
   - $m=6$ (jun.): $j\equiv 0$ → $12,24,\dots,96$.
   - $m=8$ (ago.): $j\equiv 10$ → $10,22,\dots,94$.
4. Como $q>1$, termos mais tardios são maiores. A única opção que inclui $P_{100}$ é fevereiro, então a soma paga pela empresa é máxima e o gasto do cliente, mínimo.

**Resposta:** **Fevereiro**.


```python
# ENEM Q143 — Por que comprar em FEVEREIRO minimiza o gasto?
# Modelo didático: 100 parcelas em PG crescente (q>1). A imobiliária paga
# as parcelas do mês de aniversário (maio = 5). Comparamos meses de compra.

import numpy as np
import matplotlib.pyplot as plt

# --- parâmetros do financiamento (ajuste se quiser) ---
n = 100            # número de parcelas
P1 = 1000.0        # 1ª parcela no ato da compra
r = 0.02           # acréscimo percentual fixo por mês (2%)
q = 1 + r
j = np.arange(1, n+1)
P = P1 * q**(j-1)  # P_j = P1 * q^(j-1)

# --- índices das parcelas pagas pela imobiliária (mês de aniversário = 5) ---
# regra: m + j - 1 ≡ 5 (mod 12)  ⇒  j ≡ 6 - m (mod 12)
def indices_pagos(m):
    k = (6 - m) % 12
    start = 12 if k == 0 else k
    return np.arange(start, n+1, 12)  # j = start, start+12, ...

opcoes = {2: "Fev", 4: "Abr", 5: "Mai", 6: "Jun"}

# --- soma paga pela imobiliária em cada mês de compra ---
pagos = {}
for m, nome in opcoes.items():
    idx = indices_pagos(m)
    soma = P[idx-1].sum()
    pagos[nome] = (soma, idx)

# --- economia ao escolher Fevereiro (empresa paga mais) ---
s_fev = pagos["Fev"][0]
economia = {nome: (s_fev - soma) for nome, (soma, _) in pagos.items()}

# --- gráfico 1: curva das parcelas e marcações do que a imobiliária paga ---
plt.figure(figsize=(8,4), dpi=120)
plt.plot(j, P, label="Parcela $P_j$")
for nome, (soma, idx) in pagos.items():
    plt.scatter(idx, P[idx-1], s=25, label=f"{nome}: soma paga ≈ R$ {soma:,.0f}".replace(",", "."))
plt.title("Parcelas em PG e parcelas pagas pela imobiliária")
plt.xlabel("índice da parcela $j$")
plt.ylabel("valor da parcela (R$)")
plt.grid(True, linestyle=":")
plt.legend()
plt.tight_layout()
plt.show()

# --- gráfico 2: economia ao comprar em Fevereiro vs outras opções ---
plt.figure(figsize=(6,4), dpi=120)
nomes = ["Fev", "Abr", "Mai", "Jun"]
vals = [economia[n] for n in nomes]  # Fev vira 0
plt.bar(nomes, vals)
for x, y in zip(nomes, vals):
    plt.text(x, y, f"R$ {y:,.0f}".replace(",", "."), ha="center", va="bottom")
plt.title("Economia ao escolher Fevereiro (empresa paga mais)")
plt.ylabel("diferença na soma paga pela imobiliária (R$)")
plt.grid(True, linestyle=":")
plt.tight_layout()
plt.show()

# --- resumo numérico (claro e alinhado) ---
def moeda(x): 
    return f"R$ {x:,.2f}".replace(",", ".")

print("=== Parâmetros ===")
print(f"P1 = {moeda(P1)}   |   razão q = {q:.4f}  (r = {100*r:.1f}%)")

print("\n=== Parcelas pagas pela imobiliária ===")
for nome, (soma, idx) in {k: pagos[k] for k in ["Fev","Abr","Mai","Jun"]}.items():
    # mostra primeiros 3 e a última parcela paga
    head = ", ".join(map(str, idx[:3]))
    tail = f", ..., {idx[-1]}" if len(idx) > 3 else ""
    print(f"{nome:<3}: j = [{head}{tail}]   |   soma = {moeda(soma)}")

print("\n=== Economia ao escolher Fevereiro (Fev = 0) ===")
for nome in ["Fev","Abr","Mai","Jun"]:
    print(f"{nome:<3}: {moeda(economia[nome])}")
```

- **H1 – Representações racionais:** conversão de ângulos em frações do comprimento da circunferência.

**Modelagem e solução resumida (π≈3):**
1. Trechos retos (tangentes): $\large 2\times 15=30\ \text{cm}$.
2. Arco na polia menor ($\large r=4$, $\large 150^\circ$):
   $\large
   L_{r}=\frac{150}{360}\cdot 2\pi r=\frac{5}{12}\cdot 2\cdot 3\cdot 4=10\ \text{cm}.
   $
3. Arco na polia maior ($\large R=8$, $\large 210^\circ$):
   $\large
   L_{R}=\frac{210}{360}\cdot 2\pi R=\frac{7}{12}\cdot 2\cdot 3\cdot 8=28\ \text{cm}.
   $
4. Comprimento total da correia:
   $\large
   L=30+10+28=68\ \text{cm}.
   $

**Resposta:** **$68\ \text{cm}$**.


```python
# Crescimento do comprimento de arco com o ângulo (ENEM Q144)
import numpy as np
import matplotlib.pyplot as plt

pi = 3            # aproximação usada no enunciado
r1, r2 = 4, 8     # raios das polias (cm)

theta = np.linspace(0, 360, 361)  # ângulo em graus

def arco(theta_deg, r, pi=pi):
    return (theta_deg/360) * 2 * pi * r  # s(θ) = (θ/360)·2πr

s1 = arco(theta, r1)
s2 = arco(theta, r2)

plt.figure(figsize=(7,4), dpi=120)
plt.plot(theta, s1, label=f"r = {r1} cm")
plt.plot(theta, s2, label=f"R = {r2} cm")

# marcações dos ângulos do problema
for ang in [150, 210]:
    plt.axvline(ang, linestyle=":")
    plt.scatter([ang], [arco(ang, r1)], s=40)
    plt.scatter([ang], [arco(ang, r2)], s=40)

# anotações dos valores citados no item
plt.annotate(f"{arco(150, r1):.0f} cm", xy=(150, arco(150, r1)),
             xytext=(155, arco(150, r1)+2), arrowprops=dict(arrowstyle="->"))
plt.annotate(f"{arco(210, r2):.0f} cm", xy=(210, arco(210, r2)),
             xytext=(215, arco(210, r2)+2), arrowprops=dict(arrowstyle="->"))

plt.title("Comprimento do arco s(θ) = (θ/360)·2πr  (π ≈ 3)")
plt.xlabel("Ângulo θ (graus)")
plt.ylabel("Comprimento do arco s (cm)")
plt.legend()
plt.grid(True, linestyle=":")
plt.tight_layout()
plt.show()
```

**Enquadramento (por que é C1):**
- **H3 – Operações e medidas:** uso de propriedades métricas em triângulos.
- **H2 – Padrões/propriedades:** reconhecimento de triângulo equilátero e ponto equidistante dos vértices.

**Modelagem e solução resumida:**
1. Os três postos distam **10 km entre si** ⇒ formam um **triângulo equilátero** de lado $a=10$.
2. O ponto $H$ é **equidistante** dos três vértices ⇒ é o **circuncentro** (que coincide com o baricentro no equilátero).
3. Distância do centro ao vértice $= R$ (raio da circunferência circunscrita):
   $
   R=\frac{a}{\sqrt{3}}=\frac{10}{\sqrt{3}}\approx 5{,}77\ \text{km}.
   $

**Resposta:** **entre $5$ e $6$ km**.


```python
# ENEM Q145 — Triângulo equilátero (lado 10 km) e ponto H equidistante dos vértices
import numpy as np
import matplotlib.pyplot as plt

a = 10.0  # lado (km)

# vértices do equilátero
P1 = np.array([0.0, 0.0])
P2 = np.array([a,   0.0])
P3 = np.array([a/2, (np.sqrt(3)/2)*a])

# ponto H (centro = baricentro = circuncentro no equilátero)
H = (P1 + P2 + P3) / 3

# raio da circunferência circunscrita (distância HP1 = HP2 = HP3)
R = np.linalg.norm(H - P1)

# checagem das distâncias
d1 = np.linalg.norm(H - P1)
d2 = np.linalg.norm(H - P2)
d3 = np.linalg.norm(H - P3)

print(f"Raio/circunraio R = a/√3 = {a/np.sqrt(3):.2f} km")
print(f"HP1 = {d1:.2f} km, HP2 = {d2:.2f} km, HP3 = {d3:.2f} km (iguais)")
print("Conclusão: R ≈ 5,77 km  → entre 5 e 6 km.")

# circunferência circunscrita
t = np.linspace(0, 2*np.pi, 400)
xc = H[0] + R*np.cos(t)
yc = H[1] + R*np.sin(t)

# desenho
plt.figure(figsize=(6,6))
# triângulo
plt.plot([P1[0], P2[0], P3[0], P1[0]], [P1[1], P2[1], P3[1], P1[1]])
# circunferência
plt.plot(xc, yc)
# segmentos H->vértices
for P in (P1, P2, P3):
    plt.plot([H[0], P[0]], [H[1], P[1]], linestyle=":")

# pontos e rótulos
plt.scatter(*P1); plt.annotate("P1", P1 + [ -0.3, -0.3])
plt.scatter(*P2); plt.annotate("P2", P2 + [  0.1, -0.3])
plt.scatter(*P3); plt.annotate("P3", P3 + [  0.1,  0.1])
plt.scatter(*H);  plt.annotate(f"H\nR≈{R:.2f} km", H + [0.2, 0.2])

# lado base 10 km
mid = (P1 + P2)/2
plt.annotate("10 km", xy=mid, xytext=mid + [0, -0.7], ha="center", arrowprops=dict(arrowstyle="-"))

plt.gca().set_aspect("equal", adjustable="box")
plt.title("Triângulo equilátero (a=10 km) e ponto H equidistante dos vértices")
plt.grid(True, linestyle=":")
plt.tight_layout()
plt.show()
```
